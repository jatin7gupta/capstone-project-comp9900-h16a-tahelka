The original dataset folder from https://www.kaggle.com/stefanoleone992/imdb-extensive-dataset (found in the moviedb folder) contained 4 files: 
1. IMDb movies.csv - modified below
2. IMDb names.csv - modified below
3. IMDb ratings.csv - not used in this project
4. IMDb title_principals.csv - not used in this project



# Delete columns in movieDB, replace imdb_id string with integer

```python
import pandas as pd
movies = pd.read_csv('moviedb/IMDb movies.csv', header = 0)
movies = movies.drop(['original_title', 'date_published','country','production_company','avg_vote', 'votes','usa_gross_income', 'worlwide_gross_income', 'metascore', 'reviews_from_users','reviews_from_critics', 'budget'], axis = 1)
movies['imdbid_int'] = movies.apply(lambda _: '', axis=1) #creating new column for integer version 

for i in range(len(movies)):
    movies.iloc[i,10] = int(movies.iloc[i,0][2:]) #eliminate first two 'tt' letters then convert to int

movies.to_csv('newDB/movies.csv',index=False)
movies.head()
```

# Delete columns in actor/director namesDB


```python
names = pd.read_csv('moviedb/IMDb names.csv', header = 0)
names = names.drop(['height', 'date_of_birth', 'place_of_birth', 'date_of_death','place_of_death', 'reason_of_death',  'spouses','spouses_with_children'],axis = 1)
names.to_csv('newDB/movie_filmpersonalities.csv',index=False)      #saving the file
names.head()
```

# Replace imdb_id string with integer in title_principalsDB 

```python
title_principals = pd.read_csv('moviedb/IMDb title_principals.csv', header = 0)
title_principals  = title_principals .drop(['ordering', 'job', 'characters'], axis=1)
                                                                    #creating new column for integer version 
title_principals['imdbid_int'] = title_principals.apply(lambda _: '', axis=1) 
                                                                    #creating a new column for the integer imdb_ids
for i in range(len(title_principals)):
    title_principals.iloc[i,3] = int(title_principals.iloc[i,0][2:]) #eliminate first two 'tt' letters then convert to int
    
title_principals.to_csv('newDB/title_principals.csv',index=False)    #saving the file
title_principals.head()
```
